```python

def plot_2d_graph(path,label):
    df = pd.read_csv(path)
    x = list(range(len(df)))
    y = df["Value"].values
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel("epoch_num")
    ax.set_ylabel(f"{label}")
    ax.plot(x,y)
```

```python
labels = [["bleu","BLEU4"],["lei","Levenshtein"],["loss","Loss"]]
for label in labels:
    plot_2d_graph(f"{RESULT_ROOT_PATH}/run-version_7-tag-val_{label[0]}.csv",label[1])
    
```
